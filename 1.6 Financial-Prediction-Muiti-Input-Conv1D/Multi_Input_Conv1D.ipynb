```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
print(tf.__version__)
```

```python
HIDDEN_UNITS = 200
DROPOUT_RATE = 0.5
CHANNEL1 = 128
CHANNEL2 = 256
CHANNEL3 = 512
KERNELSIZE = 3
POOLSIZE = 2

"""构建多输入模型"""
input1_= keras.Input(shape=(50, 1), name='input1')
input2_ = keras.Input(shape=(50, 1), name='input2')
input3_ = keras.Input(shape=(50, 1), name='input3')

x1 = layers.Conv1D(CHANNEL1, kernel_size=KERNELSIZE, strides=2, activation='relu', padding='same')(input1_)
x1 = layers.MaxPool1D(pool_size=POOLSIZE, strides=2)(x1)

x2 = layers.Conv1D(CHANNEL1, kernel_size=KERNELSIZE, strides=2, activation='relu', padding='same')(input2_)
x2 = layers.MaxPool1D(pool_size=POOLSIZE, strides=2)(x2)

x3 = layers.Conv1D(CHANNEL1, kernel_size=KERNELSIZE, strides=2, activation='relu', padding='same')(input3_)
x3 = layers.MaxPool1D(pool_size=POOLSIZE, strides=2)(x3)
#=================================================================================================================
x1_2 = layers.Conv1D(CHANNEL2, kernel_size=KERNELSIZE, strides=2, activation='relu', padding='same')(x1)
x1_3 = layers.MaxPool1D(pool_size=POOLSIZE, strides=2)(x1_2)

x2_2 = layers.Conv1D(CHANNEL2, kernel_size=KERNELSIZE, strides=2, activation='relu', padding='same')(x2)
x2_3 = layers.MaxPool1D(pool_size=POOLSIZE, strides=2)(x2_2)

x3_2 = layers.Conv1D(CHANNEL2, kernel_size=KERNELSIZE, strides=2, activation='relu', padding='same')(x3)
x3_3 = layers.MaxPool1D(pool_size=POOLSIZE, strides=2)(x3_2)
#=================================================================================================================
x1_4 = layers.Conv1D(CHANNEL3, kernel_size=KERNELSIZE, strides=2, activation='relu', padding='same')(x1_3)
x1_5 = layers.MaxPool1D(pool_size=POOLSIZE, strides=2)(x1_4)

x2_4 = layers.Conv1D(CHANNEL3, kernel_size=KERNELSIZE, strides=2, activation='relu', padding='same')(x2_3)
x2_5 = layers.MaxPool1D(pool_size=POOLSIZE, strides=2)(x2_4)

x3_4 = layers.Conv1D(CHANNEL3, kernel_size=KERNELSIZE, strides=2, activation='relu', padding='same')(x3_3)
x3_5 = layers.MaxPool1D(pool_size=POOLSIZE, strides=2)(x3_4)
#=================================================================================================================
x1_6 = layers.Bidirectional(layers.LSTM(HIDDEN_UNITS, return_sequences=True))(x1_5)
x1_7 = layers.Dropout(DROPOUT_RATE)(x1_6)

x2_6 = layers.Bidirectional(layers.LSTM(HIDDEN_UNITS, return_sequences=True))(x2_5)
x2_7 = layers.Dropout(DROPOUT_RATE)(x2_6)

x3_6 = layers.Bidirectional(layers.LSTM(HIDDEN_UNITS, return_sequences=True))(x3_5)
x3_7 = layers.Dropout(DROPOUT_RATE)(x3_6)
#=================================================================================================================
x1_8 = layers.Bidirectional(layers.LSTM(HIDDEN_UNITS, return_sequences=True))(x1_7)
x1_9 = layers.Dropout(DROPOUT_RATE)(x1_8)

x2_8 = layers.Bidirectional(layers.LSTM(HIDDEN_UNITS, return_sequences=True))(x2_7)
x2_9 = layers.Dropout(DROPOUT_RATE)(x2_8)

x3_8 = layers.Bidirectional(layers.LSTM(HIDDEN_UNITS, return_sequences=True))(x3_7)
x3_9 = layers.Dropout(DROPOUT_RATE)(x3_8)
#=================================================================================================================

x = layers.concatenate([x1_9, x2_9, x3_9])
x = layers.Flatten()(x)

x = layers.Dense(10, activation='relu')(x)
output_ = layers.Dense(1, activation='sigmoid', name='output')(x)

model = keras.Model(inputs=[input1_, input2_,input3_], outputs=[output_])
model.summary()
```
